```python
from aistuff import *
import time
# import time, threading
import random
import torch

# from flask import Flask, render_template, request
# import socketio
# from multiprocessing import Process, Value

# import eventlet



import pymunk
from pymunk.vec2d import Vec2d
# import pymunk.pygame_util
# global lasttime
# lasttime = time.time()
# global statetime
# statetime = 0

# print('imported')
# async_mode = 'eventlet'

# app = Flask(__name__)
# sio = socketio.Server()
# # socketapp = socketio.WSGIApp(sio)

# print(time.time())

# namespace = ""


space = pymunk.Space()  
space.gravity = 0,0
ships = {}






# @app.route('/')
# def index():
#     return render_template('index.html', namespace=namespace, async_mode=async_mode)


# @sio.on('newuser', namespace=namespace)
# def josh_test(sid, botharray):
#     print('new')
#     ships[sid] = newbox(space,80,100,botharray[0],botharray[1],sid)
    
    
#     print (botharray[0],' has joined')
#     print(sid)
#     sio.emit('identifier', sid,namespace=namespace,room=sid)
    
# #     emit('planets', )
    
# @sio.on('staterequest', namespace=namespace)
# def josh_tesat(sid,inputs):
# #     print('new')
# #     print (inputs)

#     # try:
    
#     ships[sid].input = inputs
#     AIloop()
#     SendStates()
#     # except:
#         # pass
    
    
# #     print(ships[id])

# @sio.event
# def on_connect():
#     print('new')

def newcircle(space, x,y,r):
    body = pymunk.Body(body_type=pymunk.Body.STATIC)
    body.position = (x,y)
    shape = pymunk.Circle(body,r)
    shape.elasticity =1
    shape.friction=1
    space.add(body,shape)
    return shape
manualplanets = [{'x':1000,'y':1000,'r':500}]

global planets
planets = []
for planet in manualplanets:
    planets.append(newcircle(space,planet['x'],planet['y'],planet['r']))
    


def newbox(space, size, mass,username,structure,colour):
    print('added')
    radius = Vec2d(size, size).length

    body = pymunk.Body()
    space.add(body)

    body.position = Vec2d(
        random.random()*(640 - 2*radius) + radius, 
        random.random()*(480 - 2*radius) + radius
    )
    
    shape = pymunk.Poly.create_box(body, (size, size), 0.0)
    shape.elasticity=0
    shape.friction=1
    shape.mass = mass
    shape.friction = 0.7
    shape.username = username
    shape.structure = structure
    shape.input = [False,False,False,False,False]
    
    space.add(shape)
    
    return shape

    
    
    
    
import math



def getInputs():
    p = ships["AI"].body
    
    AIship = ships["AI"]
    distance = math.hypot(p.position.x-1000,p.position.y-1000)
    movement = math.hypot(p.position.x-positionPrev['x'],p.position.y-positionPrev['y'])
    angle = p.angle
    angledif = math.atan2(p.position.x-1000,p.position.y-1000) - angle
    moveangle = math.atan2(p.velocity.y, p.velocity.x);
    
#     print(p)
    



    
    return [distance,movement,angledif,moveangle]


# global scores
# global eps_history
# scores, eps_history = [], []
struct = {0:{0:'bullet'}}
ships["AI"] = newbox(space,80,100,"OrbitAI",struct,"AI")
positionPrev = {}
positionPrev['x']=0
positionPrev['y']=0
global mindistance
mindistance = 455
# global observation

# observation = getInputs()
    
def calculatereward():
    global mindistance
    done=False
    
    AIship = ships["AI"]
    distance = math.hypot(AIship.body.position.x-1000,AIship.body.position.y-1000)
#     reward = distance/10000
    reward = 1
    movement = math.hypot(AIship.body.position.x-positionPrev['x'],AIship.body.position.y-positionPrev['y'])
    if mindistance<60:
        mindistance += 0.1
#     steps +=1
    if distance > 5000 or distance < mindistance or movement<0.01:
        done = True
        reward = -100
        mindistance=455
#         steps=0
    positionPrev['x'] = AIship.body.position.x
    positionPrev['y'] = AIship.body.position.y
    return reward,done
global episode
episode=0

score = 0


actionlist = {0:[False,True,True,False,False],
    1:[True,False,True,False,False],
    2:[False,False,True,False,False],
    3:[True,False,False,False,False],
    4:[False,False,False,False,False],
    5:[False,True,False,False,False],
    6:[True,False,False,True,False],
    7:[False,True,False,True,False],
    8:[False,False,False,True,False],
  }


def update():
    global manualplanets
    for key in ships:
        # print(ships[key].input)
        left,right,up,down,spacepress = ships[key].input
        if up:
            ships[key].body.apply_impulse_at_local_point((0,-30), (0,0))
        if left:
            ships[key].body.angular_velocity += -0.01
        else:
            if right:
                ships[key].body.angular_velocity += 0.01
        
        for planet in planets:
            # print(ships[key])
            distance = math.hypot(planet.body.position.x-ships[key].body.position.x,planet.body.position.y-ships[key].body.position.y)
            direction = math.atan2(planet.body.position.y-ships[key].body.position.y,planet.body.position.y-ships[key].body.position.x) +math.pi/2
            force = planet.area/10 / ((distance)**2)

            # print([distance,direction,force])
            
            ships[key].body.velocity += Vec2d(math.sin(direction)*force ,-math.cos(direction)*force)
        if spacepress:
            ships[key].body.position = Vec2d(1000,460)
            ships[key].body.velocity = Vec2d(0,0)
            ships[key].body.angular_velocity = 0
            ships[key].body.angle = 0
        if ships[key].body.angular_velocity > 0.3:
            ships[key].body.angular_velocity = 0.3
        else:
            if ships[key].body.angular_velocity < -0.3:
                ships[key].body.angular_velocity = -0.3
    space.step(1)

def SendStates():
    newships = {}
    for key in ships:
        oldship = ships[key]
        newships[key]={'input':oldship.input,'x':oldship.body.position.x, 'y': oldship.body.position.y, 'angle':oldship.body.angle, 'mX':oldship.body.velocity.x,'mY':oldship.body.velocity.y, 'mA':oldship.body.angular_velocity,'username':oldship.username,'colour':key,'structure':oldship.structure}
    sio.emit('states', newships, namespace=namespace)
    sio.emit('planets', manualplanets, namespace=namespace)
        
# import pygame
# pygame.init()
# screen = pygame.display.set_mode((800,800))
# clock = pygame.time.Clock()
# draw_options = pymunk.pygame_util.DrawOptions(screen)
# def pyrender():
# #     for event in pygame.event.get():
# #         if event.type == QUIT or \
# #             event.type == KEYDOWN and (event.key in [K_ESCAPE, K_q]): 
# #             exit()
#     screen.fill(pygame.color.THECOLORS["black"])
#     space.debug_draw(draw_options)
#     pygame.display.flip()
    


def AIloop():
    global episode
    global observation
    global score
    global scores
    global eps_history
    
    action = agent.choose_action(observation)
#     print(action)
    ships['AI'].input = actionlist[action]
    update()
    observation_ = getInputs()
    reward,done = calculatereward()
    score += reward
    agent.store_transition(observation, action, reward, 
                            observation_, done)
    agent.learn()
    observation = observation_
    if done:
        
        ships["AI"].input = [False,False,False,False,True]
        update()
        observation = getInputs()
        episode += 1
        done = False
        scores.append(score)
        eps_history.append(agent.epsilon)

        avg_score = np.mean(scores[-100:])

        print('episode ', episode, 'score %.2f' % score,
                'average score %.2f' % avg_score,
                'epsilon %.2f' % agent.epsilon)
        if episode % 5 == 0:
            x = [i+1 for i in range(episode)]
            filename = 'spacegraph.png'
            plotLearning(x, scores, eps_history, filename)
        score = 0

    
    
#             print(observation)
       
lasttime = time.time()
    
if __name__ == '__main__':
#     env = gym.make('LunarLander-v2')

    makenew = False
    if not makenew:
        try:
            agent = torch.load('savedagent')
            print('Loaded')
        except:
            agent = Agent(gamma=0.99, epsilon=1.0, batch_size=256, n_actions=6, eps_end=0.01,
                  input_dims=[4],eps_dec=1e-4,  lr=0.01, reptarget = 100)
            print('Made new agent')
    else:
        agent = Agent(gamma=0.99, epsilon=1.0, batch_size=256, n_actions=6, eps_end=0.01,
                  input_dims=[4],eps_dec=1e-4,  lr=0.01, reptarget = 100)
        print('Made new agent')
        

#     agent = Agent(gamma=0.99, epsilon=1.0, batch_size=256, n_actions=6, eps_end=0.01,
#               input_dims=[4],eps_dec=1e-4,  lr=0.01, reptarget = 100)
#     print('Made new agent')

    scores, eps_history = [], []
    n_games = 5000
    print(time.time())
    episode = 0
    while True:
        score = 0
        done = False
        ships["AI"].input = [False,False,False,False,True]
        update()
        observation = getInputs()
        
        while not done:
#             pyrender()
            action = agent.choose_action(observation)
            ships['AI'].input = actionlist[action]
            update()
            reward,done = calculatereward()
            observation_ = getInputs()
            score += reward
            agent.store_transition(observation, action, reward, 
                                    observation_, done)
            agent.learn()
            
            
            
            observation = observation_
        
        scores.append(score)
        eps_history.append(agent.epsilon)
        eps_history = eps_history[-20:]
        scores = scores[-20:]
        avg_score = np.mean(scores)

        print('episode ', episode, 'score %.2f' % score,
                'average score %.2f' % avg_score,
                'epsilon %.2f' % agent.epsilon, " time since ",time.time()-lasttime)
        lasttime = time.time()
        if episode % 50 == 0:
            print('saved')
            torch.save(agent,'savedagent')
        episode += 1
    torch.save(agent,'savedagent')
    print('Saved')
    x = [i+1 for i in range(n_games)]
    filename = 'lunar_lander.png'
    plotLearning(x, scores, eps_history, filename)
#     env.close()    
    
    
    
    

# if __name__ == "__main__":
#     print('a')
    
#     while True:
#         AIloop()
        
#     app = socketio.WSGIApp(sio, app)
#     eventlet.wsgi.server(eventlet.listen(('localhost', 80)), app)
    
```
